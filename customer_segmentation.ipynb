## Customer Analysis by using RFM

```python
cleanest_df = clean_df.copy()
```

```python
import datetime as dt
```

```python
snap_shot_date = cleanest_df['invoice_date_only'].max() + dt.timedelta(days= 1 )
```

```python
cleanest_df
```

```python
last_purchase = cleanest_df.groupby('Customer_Id').agg(Last_purchase = ('invoice_date_only' , 'max'))
```

```python
last_purchase['Recency'] = (snap_shot_date - last_purchase['Last_purchase']).dt.days
```

```python
last_purchase.reset_index(inplace= True)
```

Finding Frequency

```python
frequency = cleanest_df.groupby('Customer_Id')['invoice_date_only'].nunique().reset_index(name= 'Frequency')
```

```python
frequency.sort_values('Frequency' , ascending= False)
```

Finding Monetary

```python
total_revenue = cleanest_df.groupby('Customer_Id')['Revenue'].sum().reset_index()
```

```python
RFM_analysis = last_purchase.merge(frequency , on = 'Customer_Id').merge(total_revenue , on = 'Customer_Id').sort_values('Revenue' , ascending= False)
```

```python
RFM_analysis.head(20)
```

```python
r_labels = [4 , 3 , 2 , 1] # lower get the highest point
```

```python
RFM_analysis['Recency_score'] = pd.qcut(RFM_analysis['Recency'] , q = 4 , labels= r_labels)
```

```python
f_scores = [1 , 2 , 3 ]
```

We can't cut into 4 piece because more than 25 % of frequency are 1 time.

```python

RFM_analysis['F_score'] = pd.qcut(RFM_analysis['Frequency'] , q = 4 , labels= f_scores , duplicates= 'drop')
```

```python
RFM_analysis['Revenue_score'] = pd.qcut(RFM_analysis['Revenue'] , q = 4 , labels = [1 , 2 , 3 , 4])
```

```python
RFM_analysis
```

```python
def segmentation(row):
    
    ## champion customers
    if row['Recency_score'] == 4 and row['F_score'] == 3 and row['Revenue_score'] == 4:
        return 'Champions'
    
    elif row['Recency_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal Customers'
    ## new customers / proomosing
    
    elif row['Recency_score'] == 4 and row['F_score'] <= 2:
        return 'New Customers'
    
    ## Need attention and promotion almost to loose
    
    elif row['Recency_score'] <= 2 and row['F_score'] == 3 and row['Revenue_score'] >= 3:
        return 'Need Attention'
    
    ## lost customers
    
    elif row['Recency_score'] <= 2 and row['F_score'] <= 2 and row['Revenue_score'] <= 2:
        return 'Lost Customers'
    
    else:
        return 'Need Engagement'
```

```python
RFM_analysis['Segmentation'] = RFM_analysis.apply(segmentation , axis = 1)
```

```python
RFM_analysis
```

```python
RFM_summary= RFM_analysis.groupby('Segmentation').agg(average_revenue = ('Revenue' , 'mean'), average_frequecny = ('Frequency' , 'mean') , 
                                         Number_count = ('Segmentation' , 'count') , 
                                         Average_recency = ('Recency' , 'mean')).sort_values('Average_recency' , ascending= True)
```

```python
# Define churn risk using RFM SCORES (Dynamic & Adaptable)
def churn_risk(row):
    # Already Churned: Over 6 months
    if row['Recency'] > 180:
        return 'Already Churned'
    
    # High Risk: Formerly valuable customers slipping away
    # High Frequency (Top 25%) AND High Monetary (Top 25%)
    elif row['Recency'] > 90 and row['F_score'] >= 3 and row['Revenue_score'] >= 3:
        return 'High Risk (At Risk)'
    
    # Medium Risk: Customers showing early signs of disengagement
    # Medium Frequency AND Medium Monetary
    elif row['Recency'] > 60 and row['F_score'] >= 2 and row['Revenue_score'] >= 2:
        return 'Medium Risk (Need Attention)'
    
    # Active: Recent and frequent (Top 25% Frequency)
    elif row['Recency'] < 60 and row['F_score'] >= 3:
        return 'Low Risk (Active)'
    
    else:
        return 'Low Activity'
```

```python
RFM_analysis.apply(churn_risk , axis = 1).value_counts(
)
```

```python
RFM_analysis['Churn_Risk'] = RFM_analysis.apply(churn_risk , axis = 1)
```

```python
RFM_analysis[RFM_analysis['Churn_Risk'] == 'Low Activity']
```

```python
RFM_analysis[RFM_analysis['Churn_Risk'] == 'Already Churned']
```

```python
Churn_summary = RFM_analysis.groupby('Churn_Risk').agg(Number_customer = ('Customer_Id' , 'count') , 
                                                          Total_revenue = ('Revenue' , 'sum') , 
                                                          Average_revenue = ('Revenue' , 'mean'), 
                                                          Average_recency = ('Recency' , 'mean') , 
                                                          Average_frequency = ('Frequency' , 'mean'))
```
